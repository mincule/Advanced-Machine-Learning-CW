```python
args = {
    'max_depth': 100,
    'min_sample_split': 2,
    'max_features': 12
}
```

```python
param_grid = {
    'max_depth': [100, 500, 1000],
    'min_sample_split': [2, 5, 10],
    'max_features': [12, 55],
}
```

```python
grid = list(ParameterGrid(param_grid))
grid
```

```python
import time

train_time = []
inf_time = []
acc = []

for i, args in enumerate(grid):
    clf = DecisionTree(**args)
    
    # check time
    start = time.time()
    
    print(f'{i}th classifier training...')
    clf.fit(X10, y10)
    
    end1 = time.time()
    train_time.append(end1-start)
    
    y_pred = clf.predict(test_X10)
    ac = accuracy_score(test_y10, y_pred)
    
    end2 = time.time()
    inf_time.append(end2-end1)
    acc.append(ac)
```

```python
print(train_time)
print(inf_time)
print(acc)
```
