```python
from PIL import Image

import pandas as pd
import numpy as np

import torch
import torch.nn as nn
```

# 0.Loading Dataset

```python
img_train = []

for i in range(1,16009):
    ImagePATH = "data/Training/depth/"
    img_name = "depth_1_" + str(i).zfill(7) + ".png"
    ImagePATH = ImagePATH + img_name
    img_train.append(np.array(Image.open(ImagePATH)))
```

```python
img_test = []

for i in range(1,1597):
    ImagePATH = "data/Testing/depth/"
    img_name = "depth_1_" + str(i).zfill(7) + ".png"
    ImagePATH = ImagePATH + img_name
    img_test.append(np.array(Image.open(ImagePATH)))
```

```python
X_train = np.array(img_train)
X_test = np.array(img_test)

print("train dim:", X_train.shape)
print("test dim:", X_test.shape)
```
